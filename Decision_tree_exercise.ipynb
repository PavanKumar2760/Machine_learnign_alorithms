```python
import numpy as np 
import pandas as pd 

pf = pd.read_excel('Titanic_dataset_exercise.xlsx', header = [1])
pf.columns
                   
```

```python
df = pf.drop(["PassengerId", "Name","SibSp","Parch","Ticket","Cabin","Embarked"], axis = 'columns')
df
```

```python
inputs = df.drop(['Survived'], axis = "columns")
inputs
```

```python
# calculating the meadina age

median_age = df['Age'].median()
median_age
```

```python
# now I will replace all the NAN values in age column with median value

inputs['Age'] = inputs['Age'].fillna(median_age)
inputs
```

```python
output = df['Survived']
output
```

```python
# now we can see in our data set input we have some data that is text also now we need to use encoder technique 
# to convert this data in to nmbers 

from sklearn.preprocessing import LabelEncoder 

inputs['Sex'] = LabelEncoder().fit_transform(inputs['Sex'])
```

```python
inputs
```

```python
# now we will devide the data in 80% and 20 % 

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs, output, test_size = 0.15)
```

```python
# now our data has been prepared here we will import and train the model 

from sklearn import tree

model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict()  # from here we can see as per the inpiut values we will get the information about our prediction
# and this show we can corelate. 
```
