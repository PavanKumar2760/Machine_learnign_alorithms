```python

iris = df

# Extract the features and target names
X = iris.data
y = iris.target
target_names = iris.target_names

# Create a scatter plot for each feature
for feature_index in range(X.shape[1]):
    # Plot the data points for each target class
    for target_index, target_name in enumerate(target_names):
        plt.scatter(X[y == target_index, feature_index], y[y == target_index], label=target_name)

    plt.xlabel(iris.feature_names[feature_index])
    plt.ylabel('Target')
    plt.legend()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size = 0.2)
```

```python
y_test
```

```python
X_test
```

```python
# now we will train our model 

# we will import our model first 

from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix
y_predicted = model.predict(X_test)

cm = confusion_matrix(y_test, y_predicted)
```

```python
# we also can visualize the cm by seaborn 

import seaborn as sn 
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
# as our model accuracy is 100 % hence all the entris in confussion matrix will betelling the truth 
# now we will do some predictions from our model 

model.predict([[5.6, 2.8, 4.9, 2. ]])
```
