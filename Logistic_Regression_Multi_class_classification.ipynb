```python
# from here we can see that our data set have the images, data, target that we cat use to train our 
# model we can use data and target. 

# before training our model we will split the data to check the accuracy of our model.

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(df.data, df.target, test_size = .20)
```

```python
X_train
```

```python
# len(X_train)
# len(X_test)
# len(y_test)
len(y_train)
```

```python
# now as you can see that our data is ready for the training of the model hence now we will import our model 
# and will train it in the training data set 
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()


# training the model 

model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test)
```

# To check the points where  our model is failing we will draw the confusion matrix.

The confusion matrix we draw that is comparison between the data we have y_test & y_predicted that is predicted by 
the model.

```python
y_predict = model.predict(X_test)
```

```python
# importing confusion matrix 
from sklearn.metrics import confusion_matrix
# in confusion matrix the input data are Y_predicted and Y_actual 

cm = confusion_matrix (y_test, y_predict)
cm
```

```python
# we also can visualize the cm by seaborn 

import seaborn as sn 
plt.figure(figsize = (10,7))
sn.heatmap(cm,annot = True)
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
