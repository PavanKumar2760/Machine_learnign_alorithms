```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_digits
from matplotlib import pyplot as plt
digits = load_digits()
dir(digits)
```

```python
%matplotlib inline
plt.gray()
for i in range(4):
    plt.matshow(digits.images[i])
    
    
```

```python
digits.data[2]
```

```python
# Now if we make the data as data frame 
df = pd.DataFrame(digits.data)
df
```

```python
pf = pd.DataFrame(digits.target)
pf.head()
```

```python
from sklearn.model_selection import train_test_split 

X_train, X_test, y_train, y_test = train_test_split(df, pf, test_size = 0.2)
```

```python
# now we will import our model and train it in the splitted data.
# In random forest we have opportunity that 

from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
model.score(X_train, y_train)
```

```python
y_predicted = model.predict(X_test)
y_predicted
```

```python
# here we are plotting the confussion matrix 

y_predicted
Y_known = y_test

from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted)
cm
```

```python
# plottig the confussion matrix

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sn
plt.figure(figsize =(10,7))
sn.heatmap(cm, annot = True)
plt.xlabel("Predicted")
plt.ylabel("Truth")
```
